```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Maa.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

print(type(data))

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'message_date':dates})

#convert message_data type
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %H:%M - ')

df.rename(columns={'message_date':'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
# separate users and messages
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]: # username
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df['day'] = df['date'].dt.day
```

```python
df['hour'] = df['date'].dt.hour
```

```python
df['minute'] = df['date'].dt.minute
```

Finding Top 25 Words that are commonly used in chat.

# remove group messages (group notification)
# remove media omitted message
# remove all stop words

```python
# remove group messages (group notification)
temp = df[df['user'] != 'group_notification']
temp
```

```python
# remove media omitted message
temp = temp[temp['message'] != '<Media omitted>\n']
temp
```

```python
# all stop words
f = open('stop_hinglish.txt', 'r')
stop_words = f.read()
print(stop_words)
```

```python
# remove all stop words
import string
words = []

for message in temp['message']:
    for word in message.lower().split(): # convert in lower case
        # remove punctuation from each word
        word = word.translate(str.maketrans('', '', string.punctuation))
        if word not in stop_words: # remove stop words
            words.append(word)

words
```

```python
from collections import Counter
Counter(words).most_common(25)
```

```python
pd.DataFrame(Counter(words).most_common(25)).rename(columns={0:'words', 1:'frequency'}) # convert into data frame
```

```python
import emoji
```

```python
emojis = []

for message in df['message']:
    list = emoji.analyze(message)
    for emo in list:
        emojis.append(emo.chars)
emojis
```

```python
Counter(emojis).most_common()
```

```python
emoji_df = pd.DataFrame(Counter(emojis).most_common()).rename(columns={0:'emojis', 1:'frequency'})
emoji_df
```

```python
plt.pie(emoji_df['frequency'].head(), labels=emoji_df['emojis'].head(), autopct="%1.1f%%")
```

```python
plt.bar(emoji_df['emojis'].head(), emoji_df['frequency'].head())
plt.ylabel("No. of Times Emoji Used")
plt.xlabel("Emojis")
plt.show()
```
